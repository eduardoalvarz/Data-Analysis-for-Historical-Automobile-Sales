### Task 1.1: Yearly Automobile Sales Line Chart

**Description:**
Develop a line chart using pandas to illustrate the yearly fluctuations in automobile sales. The objective is to provide a clear and direct visualization of how sales vary from year to year.

**Insight:**
This chart aims to reveal patterns and trends in automobile sales over different years, allowing for a quick and insightful understanding of the historical sales performance.

- Year grouping and average calculation for 'Automobile Sales' were implemented, considering the presence of both year and month columns in the data.
- `.plot()` method was employed with `kind='line'`.
- Labels and a title were added for comprehensive visualization.

```python
# Create data for plotting
df_line = df.groupby(df['Year'])['Automobile_Sales'].mean()

# Create figure
plt.figure(figsize=(8, 4))

# Plot the line chart
df_line.plot(kind='line', marker='o', color='b')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Fluctuation of Automobile Sales Over Years')

# Show the plot
plt.show()
```

**The plot was enhanced with the following features:**

- Ticks on the x-axis were added to represent all the years, facilitating the identification of recession years.
- Annotations were included for at least two years during recession periods.
- The title was set as "Automobile Sales during Recession."

To achieve this, a list ranging from 1980 to 2023 was generated and passed to the `plt.xticks` function. Additionally, to ensure readability, the ticks on the axis were rotated at an angle. Annotations were incorporated using `plt.text(x, y, 'text to display')`.

```python
plt.figure(figsize=(10, 6))

# Create data for plotting
df_line = df.groupby(df['Year'])['Automobile_Sales'].mean()

# Mark years with recession = 1 with red points
recession_years = df[df['Recession'] == 1]['Year']
plt.scatter(recession_years, df_line.loc[recession_years], color='red', label='Recession Years', zorder=2)

# Plot the line chart
df_line.plot(kind='line', marker='o', color='blue', zorder=1)

# Set ticks on x-axis with all the years
plt.xticks(list(range(1980, 2024)), rotation=75)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Automobile Sales during Recession')

# Add grid lines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Include annotation for at least two years of recession
plt.text(1982, 650, '1981-82 Recession', fontsize=10, color='red')
plt.text(2009, 2000, '2007-09 Recession', fontsize=10, color='red')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

### Task 1.2: Recession Sales Trends by Vehicle Type

**Description:**
Plot distinct lines for vehicle types using pandas to analyze sales trends during recessions. Identify variations in sales patterns among different vehicle categories.

**Insight:**
Explore how various vehicle types are impacted during recession periods, determining if noticeable differences exist in sales trends.

```python
# Create a separate dataframe where the column 'Recession' has a value of '1'
df_recession = df[df['Recession'] == 1]

# Group by 'Year', 'Vehicle_Type' and calculate the average on 'Automobile_Sales'
df_Mline = df_recession.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Set 'Year' as index
df_Mline.set_index('Year', inplace=True)
df_Mline
```

```python
# Group by 'Vehicle_Type' and plot lines for each category
df_Mline = df_Mline.groupby(['Vehicle_Type'])['Automobile_Sales']
df_Mline.plot(kind='line', marker='o', figsize=(12, 6))

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Sales Trend Vehicle-wise during Recession')

# Add grid lines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

1. **Sports Cars:**
   - Over the years, sports cars have shown a consistent growth in sales during recession periods.
   - Despite relatively low sales, because of the high cost of the vehicle but there is a noticeable upward trend in the purchase of sports cars.

2. **Executive Cars:**
   - Executive cars had limited visibility in recession periods, with a significant peak in the year 1990.
   - In contrast to other vehicle types, executive cars experienced a distinct surge in purchases during this specific year.

3. **Medium Family Cars, Small Family Cars, and Sports Cars:**
   - Medium family cars, small family cars, and sports cars have consistently maintained popularity and were consistently purchased during recession periods.
   - These vehicle types seem to have a steady demand, showcasing a relatively stable trend over the analyzed years.

4. **Supermini Cars:**
   - Notably, the consideration for supermini cars in recession periods has declined significantly since approximately 2008.
   - Supermini cars have gradually lost their appeal as a preferred choice during recession periods.

**Inference:**

- The consistent popularity of medium family cars, small family cars, and sports cars during recession periods may be attributed to their perceived affordability and practicality, making them more resilient to economic downturns compared to luxury or niche vehicle categories.

### Task 1.4: GDP Comparison. Recession vs. NonRecession

**Description:**
Use subplots to compare GDP variations during recession and non-recession periods with line plots.

**Insight:**
Quickly visualize and compare the fluctuations in GDP, gaining insights into economic trends during both recession and stable periods.

To delve deeper into the dataset and gain insights into the observed trends in GDP during recession and non-recession periods, a comparison is made through the use of subplots. Two line charts are plotted using the `add_subplot()` function from Matplotlib, providing a visual representation of how GDP varied over time in both economic scenarios.

The goal is to answer the question:
How did the GDP fluctuate over time during recession and non-recession periods?

This comparative analysis aims to uncover patterns and potential correlations in GDP trends, contributing to a comprehensive understanding of economic dynamics.

```python
# Create dataframes for recession and non-recession period
rec_data = df[df['Recession'] == 1]
non_rec_data = df[df['Recession'] == 0]

# Figure
fig = plt.figure(figsize=(12, 6))

# Create different axes for subplotting
ax0 = fig.add_subplot(1, 2, 1)  # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2)  # add subplot 2 (1 row, 2 columns, second plot)

# Plotting for Recession Period
sns.lineplot(x='Year', y='GDP', data=rec_data, label='Recession', ax=ax0)
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')
ax0.set_title('GDP Variation during Recession Period')
ax0.set_ylim(10, 80)  # Set y-axis limit

# Plotting for Non-Recession Period
sns.lineplot(x='Year', y='GDP', data=non_rec_data, label='Non-Recession', ax=ax1)
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.set_title('GDP Variation during Non-Recession Period')
ax1.set_ylim(10, 80)  # Set y-axis limit

plt.tight_layout()
plt.show()
```

#### Inference
From this plot, it is evident that during recession, the GDP of the country was in a low range, might have afected the overall sales of the company<br><br>

### Task 1.5: Seasonality Impact Bubble Plot

**Description:**
Create a bubble plot to visually display the impact of seasonality on automobile sales.

**Insight:**
This bubble plot provides a direct representation of how seasonality influences automobile sales, offering a quick and insightful view of the seasonal variations in the sales data.

Analyzed seasonal impact on sales during non-recession years through a concise Bubble plot. Utilized `scatter()` with 'Month' and 'Automobile_Sales' as x and y, and Seasonality Weight for bubble size. Title: 'Seasonality impact on Automobile Sales'.

```python
non_rec_data = df[df['Recession'] == 0]

size = non_rec_data['Seasonality_Weight']  # for bubble effect

sns.scatterplot(data=non_rec_data, x='Month', y='Automobile_Sales', size=size, hue='Seasonality_Weight', legend=False)

plt.xlabel('Month')
plt.ylabel('Automobile_Sales')
plt.title('Seasonality impact on Automobile Sales')

plt.show()
```

#### Inference
From this plot, it is evident that seasonality has not affected on the overall sales. However, there is a drastic raise in sales in the month of April<br><br>

1. **Seasonal Promotions or Events:** April could host seasonal promotions or special events in the automotive industry, attracting more customers.

2. **Weather Conditions:** April marks the beginning of spring or warmer weather in some regions, potentially increasing interest and activity in car purchases.

3. **Tax Refunds:** In certain places, April is when people receive tax refunds, enhancing their financial capacity for significant purchases like automobiles.

4. **New Models Release:** Automotive companies often launch new models or conduct major advertising campaigns in April, stimulating higher demand.


```python
#year_data = non_rec_data[non_rec_data['Year'] == 2022]
year_data = non_rec_data[(non_rec_data['Year'] >= 1990) & (non_rec_data['Year'] <= 2020)]

size = year_data['Seasonality_Weight']  # for bubble effect

sns.scatterplot(data=year_data, x='Month', y='Automobile_Sales', size=size, hue='Seasonality_Weight', legend=False)

plt.xlabel('Month')
plt.ylabel('Automobile_Sales')
plt.title('Seasonality impact on Automobile Sales in 2022')
plt.ylim(0, 6000)  # Set y-axis limit

plt.show()
```

### Task 1.7: Pie Chart. Advertising Expenditure in Recession vs. NonRecession

**Description:**
Create a pie chart to show the proportion of advertising expenditure for XYZAutomotives during recession and non-recession periods.

**Insight:**
Quickly visualize the distribution of advertising expenditure between recession and non-recession periods for clear insights into promotional strategies.

Analyzed advertising expenditure changes for XYZAutomotives during recession and non-recession periods. Used two dataframes, calculated total Advertising_Expenditure, and visualized with `plt.pie()` titled 'Advertising Expenditure during Recession and Non-Recession Periods'.

```python
# Filter the data for recession and non-recession periods
Rdata = df[df['Recession'] == 1]
NRdata = df[df['Recession'] == 0]

# Calculate the total advertising expenditure for both periods
RAtotal = Rdata['Advertising_Expenditure'].sum()
NRAtotal = NRdata['Advertising_Expenditure'].sum()

# Create a pie chart for advertising expenditure
plt.figure(figsize=(8, 6))

labels = ['Recession', 'Non-Recession']
sizes = [RAtotal, NRAtotal]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Advertising Expenditure during Recession and Non-Recession Periods')

plt.show()
```

**Description:**
Create a map to show the company's highest sales region/offices during the recession.

**Insight:**

```python
import requests

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/us-states.json'
filename = 'us-states.json'

response = requests.get(url)
if response.status_code == 200:
    with open(filename, 'wb') as f:
        f.write(response.content)
    print(f"File '{filename}' downloaded successfully.")
else:
    print(f"Failed to download file. Status code: {response.status_code}")
```

Discovered that the dataset includes location/city information for company offices. To illustrate the recession's impact on sales across different offices/cities, a choropleth map will be developed.

```python
# Filter the data for the recession period and specific cities
recession_data = df[df['Recession'] == 1]

# Calculate the total sales by city
sales_by_city = recession_data.groupby('City')['Automobile_Sales'].sum().reset_index()

# Create a base map centered on the United States
map1 = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Create a choropleth layer using Folium
choropleth = folium.Choropleth(
    geo_data='us-states.json',  # GeoJSON file with state boundaries
    data=sales_by_city,
    columns=['City', 'Automobile_Sales'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Automobile Sales during Recession'
).add_to(map1)

# Add tooltips to the choropleth layer
choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['name'], labels=True)
)


# Display the map
map1
```
